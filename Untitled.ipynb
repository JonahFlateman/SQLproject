```python
c.execute("""SELECT reviews.reviewid, reviews.title, reviews.artist, reviews.url, reviews.score, 
          reviews.best_new_Music, reviews.author, reviews.author_type, reviews.pub_date, 
          reviews.pub_weekday, reviews.pub_day, reviews.pub_month, reviews.pub_year, 
          genres.genre, labels.label, years.year, content.content
          FROM reviews
          JOIN genres
          ON reviews.reviewid = genres.reviewid
          JOIN labels
          ON reviews.reviewid = labels.reviewid
          JOIN years
          ON reviews.reviewid = years.reviewid
          JOIN content
          ON reviews.reviewid = content.reviewid""")
df_joined = pd.DataFrame(c.fetchall())
df_joined.columns = [x[0] for x in c.description]
df_joined.head()
```

```python
df_joined = df_joined.drop_duplicates(subset=['reviewid'])
```

```python
df_joined.info()
```

```python
df_joined.author_type.value_counts()
```

```python
df_joined.author_type = df_joined.author_type.fillna('contributor')
```

```python
df_joined.genre.value_counts()
```

```python
df_joined.genre = df_joined.genre.fillna('none')
```

```python
df_joined.label = df_joined.label.fillna('none')
df_joined.year = df_joined.year.fillna('none')
```

```python
df_joined['artist'] = df_joined['artist'].str.title()
```

```python
df_best_new_music = df_joined.loc[df_joined['best_new_music'] == 1]
```

```python
df_bnm_counts = pd.DataFrame(df_best_new_music.groupby(['artist','best_new_music'])['reviewid'].count())
```

```python
df_bnm_counts.rename(columns={'reviewid':'amount'}, inplace=True)
```

```python
df_bnm_counts.sort_values(by='amount', ascending=False).reset_index()
```

```python
df_bnm_counts.value_counts()
```

```python
df_bnm_five_and_over = pd.DataFrame(df_bnm_counts.loc[df_bnm_counts['amount'].between(5, 25)].reset_index().sort_values(by='amount'))
```

What are our artists with most frequent Best New Music reviews?

```python
plt.figure(figsize=(18,8))
plt.bar(df_bnm_five_and_over['artist'], df_bnm_five_and_over['amount'])
plt.xlabel('Artist')
plt.ylabel('Number of BNM reviews')
plt.title('Most frequent BNM Artists')
plt.show()
```
